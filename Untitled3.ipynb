```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense, Normalization
import requests
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from sklearn.preprocessing import StandardScaler


```

```python
# URL of the Excel file on GitHub
excel_url = "https://github.com/NataliaGutman/AERONAUTICAL-PROJECT/raw/main/fallos.xlsx"

# Send an HTTP GET request to the URL
response = requests.get(excel_url)

if response.status_code == 200:
    # Load the Excel file into a DataFrame using pandas
    datos_df = pd.read_excel(excel_url, engine='openpyxl')
    # Now, datos_df contains the data from the Excel file.
    print(datos_df.head(10))
else:
    print("Error al recuperar datos de GitHub.")
```

```python
styled_df = datos_df.style \
    .set_properties(**{'background-color': 'lightpink', 'color': 'black'}) \

# show data frame with style ;)
styled_df
```

```python
fallos_df = datos_df.iloc[:, [0, 3, 4, 5, 6, 7, 8,9]]
print(fallos_df.head())
```

```python
X = fallos_df.iloc[:, 0:7]
y = fallos_df.iloc[:, 6]
```

```python
test_size=0.5
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y
)

```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(6, input_dim=X_train.shape[1], activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(36, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=64, validation_data=(X_test, y_test))
```

```python
predicciones = model.predict(X)
```

```python
predicciones_redondeadas = [int(round(val[0])) for val in predicciones]
```

```python
datos_df["Predicción"] = predicciones_redondeadas
```

```python
datos_df = datos_df.sort_values(by="Predicción", ascending=False)
```
